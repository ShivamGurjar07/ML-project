```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Define the text to be used for the word cloud
text = "This is some example text for the word cloud. It contains a few different words that will be used to create the word cloud."

# Create a WordCloud object with the desired parameters
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(8,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
